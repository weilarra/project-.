# 原始代码

```python
"""Backend supported: tensorflow.compat.v1, tensorflow, pytorch, paddle"""
import deepxde as dde
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
d = np.load("antiderivative_aligned_train.npz", allow_pickle=True)
X_train = (d["X"][0].astype(np.float32), d["X"][1].astype(np.float32))
y_train = d["y"].astype(np.float32)
d = np.load("antiderivative_aligned_test.npz", allow_pickle=True)
X_test = (d["X"][0].astype(np.float32), d["X"][1].astype(np.float32))
y_test = d["y"].astype(np.float32)

data = dde.data.TripleCartesianProd(
    X_train=X_train, y_train=y_train, X_test=X_test, y_test=y_test
)

# Choose a network
m = 100
dim_x = 1
net = dde.nn.DeepONetCartesianProd(
    [m, 40, 40],
    [dim_x, 40, 40],
    "relu",
    "Glorot normal",
)

# Define a Model
model = dde.Model(data, net)

# Compile and Train
model.compile("adam", lr=0.001, metrics=["mean l2 relative error"])
losshistory, train_state = model.train(iterations=10000)

# Plot the loss trajectory
dde.utils.plot_loss_history(losshistory)
plt.show()
```

# 高采样设置，200

```python
"""Backend supported: tensorflow.compat.v1, tensorflow, pytorch, paddle"""
import deepxde as dde
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

# Load dataset
d = np.load("antiderivative_aligned_train.npz", allow_pickle=True)
X_train_orig = (d["X"][0].astype(np.float32), d["X"][1].astype(np.float32))
y_train_orig = d["y"].astype(np.float32)
d = np.load("antiderivative_aligned_test.npz", allow_pickle=True)
X_test_orig = (d["X"][0].astype(np.float32), d["X"][1].astype(np.float32))
y_test_orig = d["y"].astype(np.float32)

# 重采样数据从100点到200点
original_m = 100
target_m = 200

# 定义原始和目标采样位置
x_orig = np.linspace(0, 1, original_m)
x_target = np.linspace(0, 1, target_m)

# 重采样分支网络输入（函数v）
def resample_branch_data(branch_orig, original_m, target_m):
    """重采样分支网络数据"""
    resampled = np.zeros((branch_orig.shape[0], target_m))
    for i in range(branch_orig.shape[0]):
        # 使用三次样条插值进行重采样
        interp_func = interp1d(x_orig, branch_orig[i], kind='cubic')
        resampled[i] = interp_func(x_target)
    return resampled.astype(np.float32)

# 重采样输出数据（函数u）
def resample_output_data(y_orig, original_m, target_m):
    """重采样输出数据"""
    resampled = np.zeros((y_orig.shape[0], target_m))
    for i in range(y_orig.shape[0]):
        # 使用三次样条插值进行重采样
        interp_func = interp1d(x_orig, y_orig[i], kind='cubic')
        resampled[i] = interp_func(x_target)
    return resampled.astype(np.float32)

# 对训练数据进行重采样
branch_train_resampled = resample_branch_data(X_train_orig[0], original_m, target_m)
trunk_train_resampled = x_target.reshape(-1, 1).astype(np.float32)
y_train_resampled = resample_output_data(y_train_orig, original_m, target_m)

# 对测试数据进行重采样
branch_test_resampled = resample_branch_data(X_test_orig[0], original_m, target_m)
trunk_test_resampled = x_target.reshape(-1, 1).astype(np.float32)
y_test_resampled = resample_output_data(y_test_orig, original_m, target_m)

# 创建重采样后的数据元组
X_train = (branch_train_resampled, trunk_train_resampled)
X_test = (branch_test_resampled, trunk_test_resampled)
y_train = y_train_resampled
y_test = y_test_resampled

data = dde.data.TripleCartesianProd(
    X_train=X_train, y_train=y_train, X_test=X_test, y_test=y_test
)

# Choose a network with higher sampling density (m=200)
m = 200  # 从100改为200
dim_x = 1
net = dde.nn.DeepONetCartesianProd(
    [m, 40, 40],      # 分支网络输入维度从100改为200
    [dim_x, 40, 40],  # 主干网络保持不变
    "relu",
    "Glorot normal",
)

# Define a Model
model = dde.Model(data, net)

# Compile and Train
model.compile("adam", lr=0.001, metrics=["mean l2 relative error"])
losshistory, train_state = model.train(iterations=10000)

# Plot the loss trajectory
dde.utils.plot_loss_history(losshistory)
plt.show()
```
